```python
from selenium import webdriver
from scrapy import Selector
from selenium.webdriver.chrome.service import Service
import pandas as pd

from webdriver_manager.chrome import ChromeDriverManager
import time

```

```python
driver = webdriver.Chrome(ChromeDriverManager().install())
URL = "https://www.amazon.com/All-Natural-Non-Alcoholic-Sparkling-Beverage-cranberry/product-reviews/B08DZJCQJ9/ref=cm_cr_dp_d_show_all_btm?ie=UTF8&reviewerType=all_reviews"
driver.get(URL)
```

```python
reviews = []
```

```python


sel = Selector(text=driver.page_source)

for review in sel.css('[class="a-section review aok-relative"]'):
    #print(review.css('span.a-size-base.review-text.review-text-context').getall())
    print()
    items = {
            'author':review.css('[class="a-profile-name"]::text').get(),
            'date' : review.css('[data-hook="review-date"]::text').get(),
            'review' : review.xpath('normalize-space(.//*[@data-hook="review-body"])').get(), 
            'stars' : review.css('[data-hook="review-star-rating"]').css('[class="a-icon-alt"]::text').get() 
        }
    print(items)
    reviews.append(items)
    print(len(reviews))
```

```python
df = pd.DataFrame(reviews)
```

```python
df.drop_duplicates()
```

```python
df["product_name"] = "TÖST All-Natural Non-Alcoholic Sparkling Beverage with White tea, White cranberry, Ginger, 25.4 Fl Oz (Pack of 6)"
```

```python
df.to_csv("tost reviews 2.csv")
```
