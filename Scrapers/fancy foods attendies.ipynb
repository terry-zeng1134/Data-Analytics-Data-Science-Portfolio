```python
from selenium import webdriver
from scrapy import Selector
from selenium.webdriver.chrome.service import Service
import pandas as pd

from webdriver_manager.chrome import ChromeDriverManager
import time
import random
```

```python
driver = webdriver.Chrome(ChromeDriverManager().install())
URL = "https://sff2022.mapyourshow.com/8_0/explore/exhibitor-alphalist.cfm#/"
driver.get(URL)

```

```python
driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
```

```python
sel = Selector(text=driver.page_source)

list = sel.css('[id="exhibitor-results"]').css('a::attr(href)').getall()

urls = [s for s in list if "/8_0/exhibitor" in s]

print(len(urls))

```

```python
pd.DataFrame(urls).to_csv("sff2022 urls.csv")
```

```python
brands = []
```

```python
for i in urls:
    current_page = "https://sff2022.mapyourshow.com" + i
    print(current_page)
    print(urls.index(i))
    driver.get(current_page)
    sel = Selector(text=driver.page_source)
    item = {
        "brand": sel.css('[class="ma0  mr3-m"]::text').get(),
        "url": sel.css('[title="Visit our website"]::text').get(),
        "sff2022 page": current_page
    }
    print(item)
    brands.append(item)
    time.sleep(2)
    #code to scrape website and name
```

```python
df = pd.DataFrame(brands)
```

```python
df['url'] = df['url'].str.replace("http://","")
df['url'] = df['url'].str.replace("https://","")
df['url'] = df['url'].str.replace("www.","")
#df['url'] = df['url'].map(split("/",1))[0]
df['url'] = df['url'].str.split('/').str[0]
#df['url']
```
