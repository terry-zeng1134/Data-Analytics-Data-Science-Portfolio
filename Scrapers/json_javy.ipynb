```python
import json
import pandas as pd
import urllib.request
from bs4 import BeautifulSoup
from pandas.io.json import json_normalize
import time
import requests as r
```

```python
brand_name='Javy'
brand_normalized_url='javycoffee.com'
page_num = 1
max_pages = 619  # 5 reviews per page, 3094 reviews
```

```python
import random
rev_list = []
page_num = 1
base_url = f'https://api.okendo.io/v1/stores/b0adad80-5383-4464-ba89-3c248ba60fc8/reviews?limit=5&orderBy=has_media%20desc'
while page_num <= max_pages:
    response = r.get(base_url)
    #print(base_url)
    revs = json.loads(r.get(base_url).text)['reviews']
    rev_list.append(revs)
    page_num += 1
    print("Page "+str(page_num))
    base_url = 'https://api.okendo.io/v1' +json.loads(r.get(base_url).text)['nextUrl']
    time.sleep(1+random.uniform(0,1))
```

```python
cols = ['normalized_url','brand_name','product','created_at','rev_content','review_rating']
reviews = pd.DataFrame(columns = cols)
for i in range(len(rev_list)):
    for x in range(len(rev_list[i])):
        product = rev_list[i][x]['productName']
        created_at = rev_list[i][x]['dateCreated']
        rev_content = rev_list[i][x]['body']
        rev_rating = rev_list[i][x]['rating']
        new_rev = pd.DataFrame(
                        [[brand_normalized_url, brand_name, product, created_at, rev_content, rev_rating]],
                        columns = cols)
        reviews = reviews.append(new_rev)
        
reviews = reviews.reset_index(drop = True)
```
