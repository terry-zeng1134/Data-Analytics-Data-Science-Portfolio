```python
from selenium import webdriver
from scrapy import Selector
from selenium.webdriver.chrome.service import Service
import pandas as pd

from webdriver_manager.chrome import ChromeDriverManager
import time

```

```python
driver = webdriver.Chrome(ChromeDriverManager().install())
URL = "https://fleurmarche.com/products/sex-patch"
driver.get(URL)
```

```python
num_pages = 5
reviews = []
```

```python


sel = Selector(text=driver.page_source)

for review in sel.css('[class="jdgm-rev jdgm-divider-top jdgm--done-setup"]'):
        

    items = {
            'author':review.css('[class="jdgm-rev__header"]').css('[class="jdgm-rev__author-wrapper"]').css('span.jdgm-rev__author::text').get(),
            'date' : review.css('[class="jdgm-rev__header"]').css('[class="jdgm-rev__timestamp"]::text').get(),
            'title' : review.css('[class="jdgm-rev__content"]').css('[class="jdgm-rev__title"]::text').get(),
            'review' : review.css('[class="jdgm-rev__content"]').css('[class="jdgm-rev__body"]').css('p::text').get(), 
            'stars' : review.css('[class="jdgm-rev__header"]').css('[class="jdgm-rev__rating"]').attrib['data-score'], 
        }
    print(items)
    reviews.append(items)
    print(len(reviews))
```

```python
df = pd.DataFrame(reviews)
```

```python
df.drop_duplicates()
```

```python
df.to_csv("fleur marche reviews.csv")
```
