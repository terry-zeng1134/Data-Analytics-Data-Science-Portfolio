```python
from selenium import webdriver
from scrapy import Selector
from selenium.webdriver.chrome.service import Service
import pandas as pd

from webdriver_manager.chrome import ChromeDriverManager
import time
import random
```

```python
driver = webdriver.Chrome(ChromeDriverManager().install())
URL = "https://athleticgreens.com/reviews/"
driver.get(URL)
num_pages = 836
start_page = 1
reviews = []
```

```python
for i in range(start_page, num_pages+1):
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    
    sel = Selector(text=driver.page_source)
    first_item = {
            'author':sel.css('[class="yotpo-review yotpo-regular-box yotpo-regular-box-filters-padding "]').css('[class="y-label yotpo-user-name yotpo-font-bold pull-left"]::text').get(),
            'date' : sel.css('[class="yotpo-review yotpo-regular-box yotpo-regular-box-filters-padding "]').css('[class="y-label yotpo-review-date"]::text').get(),
            'review' : sel.css('[class="yotpo-review yotpo-regular-box yotpo-regular-box-filters-padding "]').css('[class="content-review"]::text').get(), 
            'stars' : sel.css('[class="yotpo-review yotpo-regular-box yotpo-regular-box-filters-padding "]').css('[class="sr-only"]::text').get()[0:1]     
    }
    reviews.append(first_item)
    for review in sel.css('[class="yotpo-review yotpo-regular-box  "]'):
        items = {
            'author':review.css('[class="y-label yotpo-user-name yotpo-font-bold pull-left"]::text').get(),
            'date' : review.css('[class="y-label yotpo-review-date"]::text').get(),
            'review' : review.css('[class="content-review"]::text').get(),
            'stars' : review.css('[class="sr-only"]::text').get()[0:1] 
        }
        if review.css('[class="rest-content-collapsed"]::text').get():
            items['review'] += review.css('[class="rest-content-collapsed"]::text').get()
        reviews.append(items)
        
    print("Review count: " + str(len(reviews)))
    time.sleep(5+random.uniform(0,5))
    driver.find_element_by_css_selector('[data-page="{page}"]'.format(page=i)).click()
    print("Page " + str(i) )
```
