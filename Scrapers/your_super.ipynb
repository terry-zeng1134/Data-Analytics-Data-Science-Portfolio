```python
brand_name='Your Super'
brand_normalized_url='yoursuper.com'
rev_list = []
product = 1
```

```python
for i in api_url:
    print("Product number " + str(product))
    print(url_dict[i])
    product+=1
    current_url = i
    not_done = True
    while not_done:
        try:
            revs = json.loads(r.get(current_url).text)['reviews']
            rev_list.append(revs)
            next_url = json.loads(r.get(current_url).text)['nextUrl']
            current_url = 'https://api.okendo.io/v1' + next_url
            time.sleep(1+random.uniform(0,1))
            print("Review count: " + str(len(rev_list)))
        except KeyError:
            not_done = False

   
```

```python
cols = ['normalized_url','brand_name','product','created_at','rev_content','review_rating']
reviews = pd.DataFrame(columns = cols)
for i in range(len(rev_list)):
    for x in range(len(rev_list[i])):
        product = rev_list[i][x]['productName']
        created_at = rev_list[i][x]['dateCreated']
        rev_content = rev_list[i][x]['body']
        rev_rating = rev_list[i][x]['rating']
        new_rev = pd.DataFrame(
                        [[brand_normalized_url, brand_name, product, created_at, rev_content, rev_rating]],
                        columns = cols)
        reviews = reviews.append(new_rev)
        
reviews = reviews.reset_index(drop = True)
```
