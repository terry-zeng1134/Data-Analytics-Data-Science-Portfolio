```python
# Imports
import numpy as np
import pandas as pd
import nltk
import contractions
import math
import re
import string
from google.cloud import bigquery
```

```python
client = bigquery.Client(location="US")
print("Client creating using default project: {}".format(client.project))

client = bigquery.Client(location="US", project="helio-staging")
```

```python
parent_cat = ["personal care & beauty",
                   
                    "Personal Care & Beauty",
                    ]
```

```python
job_config = bigquery.QueryJobConfig(
        query_parameters=[
            bigquery.ArrayQueryParameter("parent_cat", "STRING", parent_cat)
        ]
)
max_date_query = """
    SELECT normalized_url, brand_attribute, parent_category, primary_category 
    FROM `helio-staging.attributes.brand_attributes_cleaned__1_0` 
    WHERE normalized_url = 'freckbeauty.com'
    AND parent_category in UNNEST (@parent_cat)
    """
query_job = client.query(
    max_date_query,
    # Location must match that of the dataset(s) referenced in the query.
    location="US",
    job_config=job_config
)  # API request - starts the query

attributes = query_job.to_dataframe()
attributes = attributes.loc[attributes.astype(str).drop_duplicates().index]
```

```python
freck_attributes = attributes['brand_attribute'].tolist()
```
